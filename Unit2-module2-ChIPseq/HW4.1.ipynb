#### 2.2 For files in Sample_fastq_trimmed, align all to chr12 into sam files,  and compile all the stderr to a file named all.align.out

```python
%%bash
cd sample_fastq_trimmed/
BT_INDEX="../UCSC_mm10/Bowtie2Index_mm10chr12/genome_chr12"
for file in *.fq
do
    bowtie2 -x $BT_INDEX -U $file -S "$file.sam" &> "$file.align.out"
done
```

```python
%%bash
cd sample_fastq_trimmed/
OUTPUT="all_in_all.out"
for file in *.align.out
do
    echo "$file:" >> $OUTPUT
    cat $file >> $OUTPUT
    echo " " >> $OUTPUT
    echo "----------------" >> $OUTPUT
    echo " " >> $OUTPUT
done

cat $OUTPUT

mv $OUTPUT "./all.align.out"
```

### 3

#### 3.1 convert to bam, sort and filter

```python
%%bash
cd sample_fastq_trimmed/
for file in *.sam
do
    filename=${file%%.*}
    # Convert sam file to bam file
    samtools view -bS $file > $filename.bam
    # Sort bam file
    samtools sort $filename.bam -o $filename.srt.bam
    # Remove PCR duplicates
    samtools rmdup -S $filename.srt.bam $filename.srt.dupr.bam
done
```

#### 3.2 Summarize the sizes for all the sam files / bam files / sorted bam files / sorted and filtered bam files into a txt file named sum_sizes.txt

```python
%%bash
cd sample_fastq_trimmed/

OUTPUT="sum_sizes.txt"

printf "name\tsize\n" >> $OUTPUT

ext='.500_trimmed.fq.sam .bam .srt.bam'

for file in *.sam
do
    genename=${file%%.*}
    for e in $ext
    do
        ls -lh "$genename$e" | awk {'print $9, "\t" ,$5'} >> $OUTPUT
    done
done

cat $OUTPUT
```
